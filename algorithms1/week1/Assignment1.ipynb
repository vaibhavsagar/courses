```python
'''
import Data.List (sort)

countSortedInversions ll lr (lm, acc) = case (ll, lr) of
    ([], _)        -> ((lm ++ sort lr), acc)
    (_, [])        -> ((lm ++ sort ll), acc)
    (hl:tl, hr:tr)
        | hl <= hr -> countSortedInversions tl lr (lm ++ [hl], acc)
        | hr <  hl -> countSortedInversions ll tr (lm ++ [hr], acc + length ll)

countInversions list
    | length list < 2  = (list, 0)
    | length list == 2 = let (h:t:[]) = list in
        if h > t 
            then (t:[h], 1)
            else (list,  0)
    | otherwise        = let
        half = (length list) `div` 2
        il@(ll, cl) = countInversions $ take half list
        ir@(lr, cr) = countInversions $ drop half list
        im@(lm, cm) = countSortedInversions ll lr ([], 0)
        in (lm, cl + cr + cm)

snd $ countInversions array
'''
```

```python
def count_sorted_inversions(ll, lr, accm):
    (lm, acc) = accm
    if len(ll) == 0:
        return (list(sorted(lm + lr)), acc)
    elif len(lr) == 0:
        return (list(sorted(lm + ll)), acc + len(ll))
    else:
        (hl, *tl) = ll
        (hr, *tr) = lr
        if hl <= hr:
                return count_sorted_inversions(tl, lr, (lm+[hl], acc))
        else:
                return count_sorted_inversions(ll, tr, (lm+[hr], acc + len(ll)))

def count_inversions(ls):
    if len(ls) < 2: return (ls, 0)
    elif len(ls) == 2:
        (h, t, *rest) = ls
        if h > t:
            return (t + [h], 1)
        else:
            return (ls, 0)
    else:
        half = len(ls) // 2
        (ll, cl) = count_inversions(ls[0:half])
        (lr, cr) = count_inversions(ls[half:])
        (lm, cm) = count_sorted_inversions(list(sorted(ll)), list(sorted(lr)), ([], 0))
        return (lm, cl+cr+cm)

with open("IntegerArray.txt") as fl:
    content = list(map(int, fl.read().split()))
    count_inversions(content)
```
